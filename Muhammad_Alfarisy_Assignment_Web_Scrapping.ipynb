```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL halaman daftar buku di Project Gutenberg
url = "https://www.gutenberg.org/ebooks/search/?query=fiction"

# Mengirim permintaan GET ke situs web
response = requests.get(url)

# Memeriksa apakah permintaan berhasil
if response.status_code == 200:
    # Parsing konten HTML
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Menemukan elemen buku
    book_elements = soup.find_all('li', class_='booklink')
    
    # Daftar untuk menyimpan data buku
    books = []
    
    # Loop melalui elemen buku dan ekstrak informasi
    for book in book_elements:
        title = book.find('span', class_='title').text.strip()  # Judul buku
        author = book.find('span', class_='author').text.strip() if book.find('span', class_='author') else 'Unknown'  # Penulis
        link = "https://www.gutenberg.org" + book.find('a')['href']  # Link ke halaman buku
        
        # Tambahkan data buku ke daftar
        books.append({
            'Title': title,
            'Author': author,
            'Link': link
        })
    
    # Konversi daftar ke DataFrame
    df = pd.DataFrame(books)
     
    # Menyimpan data ke CSV
    df.to_csv('gutenberg_books.csv', index=False)
    print("Data telah disimpan ke 'gutenberg_books.csv'.")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")

df    
```
